```python
import requests
from bs4 import BeautifulSoup
import random
def scrapeURLs(url):
    listOfLinks = []
    response = requests.get(
        url=url,
    )
    soup = BeautifulSoup(response.content, 'html.parser')

    title = soup.find(id="firstHeading")

    # Get all the links
    allLinks = soup.find(id="bodyContent").find_all("a")
    linkToScrape = 0

    for link in allLinks:
        # We are only interested in other wiki articles
        if link['href'].find("/wiki/") == -1: 
            continue

        linkToScrape = link
        listOfLinks.append(str("https://en.wikipedia.org"+linkToScrape['href']))
    return(listOfLinks)
```

```python
listOfLinksToScrape = scrapeURLs("https://en.wikipedia.org/wiki/List_of_tourist_attractions_in_Denmark#Top_50_most_visited_attractions_in_Denmark")
```

```python
def scrapeInfo(url):
    listOfLinks = []
    response = requests.get(
        url=url,
    )
    soup = BeautifulSoup(response.content, 'html.parser')
    title = soup.find(id="firstHeading")
    allLinks = soup.find(id="bodyContent").find_all("a")
    linkToScrape = 0

    for link in allLinks:
        # We are only interested in other wiki articles
        if link['href'].find("/wiki/") == -1: 
            continue

        linkToScrape = link
        listOfLinks.append(str("https://en.wikipedia.org"+linkToScrape['href']))
    return(listOfLinks)
```

```python
def conversion(old):
    direction = {'N':1, 'S':-1, 'E': 1, 'W':-1}
    new = old.replace(u'°',' ').replace('′',' ').replace('″',' ')
    new = new.split()
    new_dir = new.pop()
    new.extend([0,0,0])
    return (int(new[0])+int(new[1])/60.0+int(new[2])/3600.0) * direction[new_dir]
```

```python
import wikipedia
import pandas as pd
import os
import re

ShortDescription = []
LongDescription = []
latitude_list = []
longitude_list = []
df_adventures = pd.DataFrame()

for name in names:
    sDesc = wikipedia.summary(name, sentences=1)
    sDesc = re.sub('\.(?!\s|\d|$)', '. ', sDesc)
    lDesc = wikipedia.summary(name, sentences=5)
    lDesc = lDesc.replace(';','')
    lDesc = re.sub('== [^>]+ ==', '', lDesc)
    lDesc = os.linesep.join([s for s in lDesc.splitlines() if s])
    lDesc = str.join(" ", lDesc.splitlines())
    print(sDesc + '\n')
    ShortDescription.append(sDesc)
    LongDescription.append(lDesc)

    try:
        req = requests.get("https://en.wikipedia.org/wiki/"+name).text
        soup = BeautifulSoup(req, 'lxml')
        
        latitude_old = soup.find("span", {"class": "latitude"}).text
        longitude_old = soup.find("span", {"class": "longitude"}).text
        latitude_list.append(conversion(latitude_old))
        longitude_list.append(conversion(longitude_old))
    except:
        latitude_list.append('0')
        longitude_list.append('0')
df_adventures['Title'] = names
df_adventures['ShortDescription'] = ShortDescription
df_adventures['LongDescription'] = LongDescription
df_adventures['Category'] = df_adventures.Title.map(manual_categories)
df_adventures['Latitude'] = latitude_list
df_adventures['Longitude'] = longitude_list
df_adventures['Price'] = df_adventures.Title.map(manual_prices)
df_adventures['Picture'] = 0
```

```python
df_adventures_r = df_adventures[(df_adventures[['Latitude','Longitude']] != '0').all(axis=1)]
df_adventures_r= df_adventures_r.reset_index(drop=True)
df_adventures_r
```
